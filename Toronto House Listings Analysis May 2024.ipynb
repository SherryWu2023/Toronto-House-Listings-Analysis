```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
import pandas as pd
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = "https://www.realtor.ca/on/toronto/real-estate"
driver.get(url)

# Wait for the cards to be visible
wait = WebDriverWait(driver, 30)
cards = wait.until(EC.visibility_of_all_elements_located((By.XPATH, "//li[contains(@class,'cardCon')]/div/a")))

# List to store page URLs
page_urls = [card.get_attribute('href') for card in cards]

# Data list to collect all the information
data_list = []

# Loop through each page URL
for page_url in page_urls:
    if page_url:
        driver.get(page_url)
        wait.until(EC.presence_of_element_located((By.XPATH, "//div[contains(@id,'listingPriceValue')]")))

        
        try:
            price = driver.find_element(By.XPATH, "//div[contains(@id,'listingPriceValue')]").text
            address = driver.find_element(By.XPATH, "//h1[contains(@id,'listingAddress')]").text
            bedrooms = driver.find_element(By.XPATH, '//*[@id="BedroomIcon"]/div[2]').text
            bathrooms = driver.find_element(By.XPATH, '//*[@id="BathroomIcon"]/div[2]').text
            
            # Store data in a dictionary
            data = {
                "price": price,
                "address": address,
                "rooms": bedrooms,
                "bathrooms": bathrooms
            }
            data_list.append(data)
            
        except Exception as e:
            print(f"Error scraping data for {page_url}: {e}")



# Print collected data
for data in data_list:
    print(data)

```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Target URL
url = "https://www.realtor.ca/on/toronto/real-estate"
driver.get(url)

wait = WebDriverWait(driver, 30)
data_list = []

current_page = 1
last_page = 25  

#the xpath to the same element varies from page to page ,so a list of variations is created
xpaths = [
    "//li[contains(@class,'cardCon')]/div/a",  
    "//div[contains(@class,'cardCon')]/div/a"  
]

while True:  
    cards_found = False
    for xpath in xpaths:
        try:
            
            cards = wait.until(EC.visibility_of_all_elements_located((By.XPATH, xpath)))
            if cards:
                cards_found = True
                break  
        except TimeoutException:
            continue  # Try the next XPath if current one fails

    if not cards_found:
        print("Failed to find cards with any known XPath.")
        break

    # List to store page URLs
    page_urls = [card.get_attribute('href') for card in cards]

    # Process each page URL
    for page_url in page_urls:
        if page_url:
            
            driver.get(page_url)
            wait.until(EC.presence_of_element_located((By.XPATH, "//div[contains(@id,'listingPriceValue')]")))
            
            try:
                price = driver.find_element(By.XPATH, "//div[contains(@id,'listingPriceValue')]").text
                address = driver.find_element(By.XPATH, "//h1[contains(@id,'listingAddress')]").text
                bedrooms = driver.find_element(By.XPATH, '//*[@id="BedroomIcon"]/div[2]').text
                bathrooms = driver.find_element(By.XPATH, '//*[@id="BathroomIcon"]/div[2]').text

                data = {
                    "price": price,
                    "address": address,
                    "bedrooms": bedrooms,
                    "bathrooms": bathrooms
                }
                data_list.append(data)
            except Exception as e:
                print(f"Error scraping data for {page_url}: {e}")
            finally:
                driver.back()  # Navigate back to the listing page

    if current_page >= last_page:
        print(f"Reached the last page: {last_page}. Ending scrape.")
        break

    try:
        next_page_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "#ListViewPagination_Bottom .lnkNextResultsPage div")))
        driver.execute_script("arguments[0].scrollIntoView();", next_page_button)
        next_page_button.click()
        current_page += 1
    except TimeoutException:
        print("No more pages or failed to click next.")
        break

driver.quit()

df2 = pd.DataFrame(data_list)
df2.to_csv('Toronto_listings_2.csv', index=False)
print("Data collection complete and saved to CSV.")
```

```python
# Function to clean addresses
def clean_address(address):
    parts = address.strip().split('\n')
    cleaned_address = ', '.join(parts[:-1])  # Removes the postal code
    return cleaned_address

# Load data
df2 = pd.read_csv('Toronto_listings_2.csv')

# Apply the cleaning function
df2['cleaned_address'] = df2['address'].apply(clean_address)
df2['cleaned_address'] = df2['cleaned_address'] + ', Toronto, Ontario'

# Print results for verification
print(df2[['address', 'cleaned_address']].head())

# Save to new CSV
df2.to_csv('toronto_cleaned.csv', index=False)
```

```python
from geopy.geocoders import Nominatim
import time
from geopy.exc import GeocoderTimedOut, GeocoderQuotaExceeded
import pandas as pd

def geocode_address(geolocator, address, retry_count=0):
    try:
        # Attempt to geocode the address
        location = geolocator.geocode(address, timeout=10)
        if location:
            return (location.latitude, location.longitude)
        else:
            return (None, None)
    except GeocoderTimedOut:
        if retry_count < 3:  # Limit retries to 3
            time.sleep(5)  # Wait for a second and retry
            return geocode_address(geolocator, address, retry_count+1)
        else:
            return (None, None)
    except GeocoderQuotaExceeded:
        # Handle quota exceeded error
        print("Quota exceeded, please try again later")
        return (None, None)

# Load your DataFrame
df2 = pd.read_csv('toronto_cleaned.csv')

# Define a unique user-agent
geolocator = Nominatim(user_agent="torontosherry")

# Apply geocode_address to each row in the DataFrame
df2['coordinates'] = df2['cleaned_address'].apply(lambda x: geocode_address(geolocator, x))

# Split coordinates into separate columns
df2[['latitude', 'longitude']] = pd.DataFrame(df2['coordinates'].tolist(), index=df2.index)

# Print result or save to CSV
print(df2.head())
df2.to_csv('geocoded_addresses_toronto.csv', index=False)
```

```python
#Clean the bedrooms column
import pandas as pd

def sum_bedrooms(bedroom_string):
    numbers = map(int, bedroom_string.split('+'))
    return sum(numbers)

df4=pd.read_csv('geocoded_addresses_toronto.csv')
df4['bedrooms']=df4['bedrooms'].apply(sum_bedrooms)
```
